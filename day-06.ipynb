## Question 2

Create a pivot table to summarize the total sales volume of ice cream products by month and temperature range.
Use the following temperature bins where each bin excludes the upper bound but includes the lower bound:
- Less than 60 degrees
- 60 to less than 70 degrees
- 70 to less than 80 degrees
- 80 to less than 90 degrees
- 90 to less than 100 degrees
- 100 degrees or more

```python
import pandas as pd

# Define bins and labels
bins = [-float('inf'), 60, 70, 80, 90, 100, float('inf')]
labels = [
    "<60°F",
    "60–69°F",
    "70–79°F",
    "80–89°F",
    "90–99°F",
    ">=100°F"
]

# Assign temperature range
ice_cream_sales_data['temp_range'] = pd.cut(
    ice_cream_sales_data['temperature'],
    bins=bins,
    labels=labels,
    right=False   # excludes upper bound
)

# Extract month (as YYYY-MM)
ice_cream_sales_data['month'] = ice_cream_sales_data['sale_date'].dt.to_period('M').astype(str)

# Pivot: total sales_volume by month & temp_range
pivot_table = pd.pivot_table(
    ice_cream_sales_data,
    values='sales_volume',
    index='month',
    columns='temp_range',
    aggfunc='sum',
    fill_value=0
).reset_index()

print(pivot_table)


"""
#SQL Logic
WITH sales_with_bins AS (
    SELECT
        DATE_TRUNC('month', sale_date)::date AS month,
        CASE 
            WHEN temperature < 60 THEN '<60°F'
            WHEN temperature >= 60 AND temperature < 70 THEN '60–69°F'
            WHEN temperature >= 70 AND temperature < 80 THEN '70–79°F'
            WHEN temperature >= 80 AND temperature < 90 THEN '80–89°F'
            WHEN temperature >= 90 AND temperature < 100 THEN '90–99°F'
            ELSE '>=100°F'
        END AS temp_range,
        sales_volume
    FROM ice_cream_sales_data
)
SELECT 
    month,
    temp_range,
    SUM(sales_volume) AS total_sales_volume
FROM sales_with_bins
GROUP BY month, temp_range
ORDER BY month, temp_range;
"""
```

## Question 3

Can you detect any outliers in the monthly sales volume using the Inter Quartile Range (IQR) method? A month is considered an outlier if falls below Q1 minus 1.5 times the IQR or above Q3 plus 1.5 times the IQR.

```python
# Ensure sale_date is datetime
ice_cream_sales_data['sale_date'] = pd.to_datetime(ice_cream_sales_data['sale_date'])

# Extract month (as YYYY-MM)
ice_cream_sales_data['month'] = ice_cream_sales_data['sale_date'].dt.to_period('M')

# Aggregate monthly total sales
monthly_sales = (
    ice_cream_sales_data
    .groupby('month')
    .agg(total_sales=('sales_volume','sum'))
    .reset_index()
)

# IQR calculation
Q1 = monthly_sales['total_sales'].quantile(0.25)
Q3 = monthly_sales['total_sales'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
higher_bound = Q3 + 1.5 * IQR

# Detect outliers
outliers = monthly_sales[
    (monthly_sales['total_sales'] < lower_bound) |
    (monthly_sales['total_sales'] > higher_bound)
]

# Print results
print(f"Lower Bound: {lower_bound}, Upper Bound: {higher_bound}")
print("Monthly Sales:")
print(monthly_sales)
print("Outliers:")
print(outliers)

"""
# Boxplot for visualization 
plt.figure(figsize=(8,5))
plt.boxplot(monthly_sales['total_sales'], vert=False)
plt.title("Monthly Ice Cream Sales Volume (with Outliers)")
plt.xlabel("Sales Volume")
plt.show()

#SQL logic
WITH monthly_sales AS (
  SELECT 
    DATE_TRUNC('month', sale_date) AS month,
    SUM(sales_volume) AS total_sales
  FROM ice_cream_sales_data
  GROUP BY DATE_TRUNC('month', sale_date)
),
percentiles AS (
  SELECT 
    PERCENTILE_CONT(0.25) WITHIN GROUP (ORDER BY total_sales) AS Q1,
    PERCENTILE_CONT(0.75) WITHIN GROUP (ORDER BY total_sales) AS Q3
  FROM monthly_sales
),
bounds AS (
  SELECT 
    Q1,
    Q3,
    (Q3 - Q1) AS IQR,
    Q1 - 1.5 * (Q3 - Q1) AS lower_bound,
    Q3 + 1.5 * (Q3 - Q1) AS upper_bound
  FROM percentiles
)
SELECT 
  m.month,
  m.total_sales,
  CASE 
    WHEN m.total_sales < b.lower_bound THEN 'Low Outlier'
    WHEN m.total_sales > b.upper_bound THEN 'High Outlier'
    ELSE 'Normal'
  END AS outlier_flag
FROM monthly_sales m
CROSS JOIN bounds b
ORDER BY m.month;

"""
```
