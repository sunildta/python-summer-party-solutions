## Question 2

Using the cleaned data, calculate the total pre-sale orders per month for each region and demographic group.

```python
#drop missing values
cleaned_pre_sale_data = pre_sale_data.dropna()
#convert to datetime for time operation
cleaned_pre_sale_data['pre_order_date'] = pd.to_datetime(cleaned_pre_sale_data['pre_order_date'] )
#extract year month as we need monthly total_pre_sale orders
cleaned_pre_sale_data['month'] = cleaned_pre_sale_data['pre_order_date'].dt.to_period('M')

#group by region,month,demographic_group
pre_sale_month = cleaned_pre_sale_data.groupby(['month','region','demographic_group']
                                              ).agg(total_order=('pre_order_quantity','sum')).reset_index()
#convert month back to string
pre_sale_month['month'] = pre_sale_month['month'].astype(str)

print(pre_sale_month)




"""
SQL query
SELECT 
  SUM(pre_order_quantity) AS total_order,
  EXTRACT(YEAR FROM pre_order_date) AS year,
  EXTRACT(MONTH FROM pre_order_date) AS month,
  region,
  demographic_group
FROM pre_sales_data
GROUP BY 
  EXTRACT(YEAR FROM pre_order_date),
  EXTRACT(MONTH FROM pre_order_date),
  region,
  demographic_group
ORDER BY year, month, region, demographic_group;
"""
```

## Question 3

Predict the total pre-sales quantity for each region for September 2024. Assume that growth rate from August to September, is the same as the growth rate from July to August in each region.

```python
#So this Nike team want me to predict what percentage of total pre-sales quantity will be in September with the help of its previous month growth
#drop missing values
cleaned_pre_sale_data = pre_sale_data.dropna()

#convert to datetime for time operation
cleaned_pre_sale_data['pre_order_date'] = pd.to_datetime(cleaned_pre_sale_data['pre_order_date'] )

#extract year month as we need monthly total_pre_sale orders
cleaned_pre_sale_data['month'] = cleaned_pre_sale_data['pre_order_date'].dt.to_period('M')

#july to august pre_sale_data
pre_sale_month = cleaned_pre_sale_data.groupby(['month','region']).agg(total_order=('pre_order_quantity','sum')).reset_index()
july_august = pre_sale_month[pre_sale_month['month'].isin([pd.Period('2024-07'),pd.Period('2024-08')])]
pivot_jul_aug = july_august.pivot(index='region',columns='month',values='total_order').reset_index()

pivot_jul_aug['growth_rate'] = (pivot_jul_aug[pd.Period('2024-08')] - pivot_jul_aug[pd.Period('2024-07')]) / pivot_jul_aug[pd.Period('2024-07')]

#  Predict September = August * (1 + growth_rate)
pivot_jul_aug['predicted_sep_2024'] = pivot_jul_aug[pd.Period('2024-08')] * (1 + pivot_jul_aug['growth_rate'])

print(pivot_jul_aug[['region', 'predicted_sep_2024']])





"""
SQL logic:
WITH july_to_august as
(
  SELECT 
    DATE_TRUNC('Month',pre_order_date)::date as month_start,
    region,
    SUM(pre_order_quantity) as total_order
  FROM pre_sale_data
  WHERE pre_order_date>='2024-07-01' AND pre_order_date<'2024-09-01'
  GROUP BY DATE_TRUNC('Month',pre_order_date),region
),
growth_calc as (
  SELECT 
  region,
  MAX(CASE WHEN month_start='2024-07-01' THEN total_order END) as july_sales,
  MAX(CASE WHEN month_start='2024-08-01' THEN total_order END) as august_sales
  FROM july_to_august 
  GROUP BY region
)
SELECT region,
 region,
 (august_sales * (1 + (august_sales - july_sales)::decimal / july_sales)) AS predicted_sep_2024
FROM growth_calc;
"""
```
